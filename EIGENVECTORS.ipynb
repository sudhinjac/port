```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import yfinance as yf

# Step 1: Download historical stock data for Apple, Google, Meta, and Juniper
tickers = ['TANLA.NS', 'HCLTECH.NS', 'NESCO.NS', 'NESTLEIND.NS']

# Download data from Yahoo Finance (past 5 years)
data = yf.download(tickers, start='2021-01-01')['Adj Close']

# Step 2: Calculate daily returns
returns = data.pct_change().dropna()

# Step 3: Perform PCA on the stock returns
pca = PCA()
pca.fit(returns)

# Get the eigenvalues (variance explained) and eigenvectors (principal components)
eigenvalues = pca.explained_variance_
eigenvectors = pca.components_

# Step 4: Explained variance ratio (variance explained by each principal component)
explained_variance_ratio = pca.explained_variance_ratio_

# Step 5: Cumulative Explained Variance Plot
plt.figure(figsize=(8, 6))
plt.plot(np.cumsum(explained_variance_ratio), marker='o', linestyle='--', color='b')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()

# Step 6: Display the eigenvalues and eigenvectors
print("Eigenvalues (Variance Explained by Each Component):")
print(eigenvalues)

print("\nEigenvectors (Principal Components):")
print(eigenvectors)

print("\nExplained Variance Ratio (Proportion of Variance Explained by Each Component):")
print(explained_variance_ratio)
```
