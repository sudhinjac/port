```python
import numpy as np
import pandas as pd
from scipy.stats import norm
import pandas_datareader.data as web
import datetime
import math

class ValueAtRiskMonteCarlo:
    
    def __init__(self, S, mu, sigma, c, n, iterations):
        self.S = S
        self.mu = mu
        self.sigma = sigma
        self.c = c
        self.n = n
        self.iterations = iterations 

    def simulation(self):
        stock_data = np.zeros([self.iterations, 1])	
        rand = np.random.normal(0, 1, [1, self.iterations])

        #equation for the S(t) stock price
        stock_price = self.S*np.exp(self.n*(self.mu - 0.5*self.sigma**2)+self.sigma*np.sqrt(self.n)*rand)
        #we have to sort the stock prices to determine the percentile
        stock_price = np.sort(stock_price)

        #it depends on the confidence level: 95% -> 5 and 99% -> 1
        percentile = np.percentile(stock_price,(1-self.c)*100)

        return self.S-percentile
```

```python

# Initialize DataFrame
df = pd.DataFrame(columns=['Ticker', 'Cov_mar', 'Market_var', 'Beta', 'Volatility%', 'Return%', 'CV', 'Sharp Ratio',
                           'CAGR', 'MAXDD%', 'Value at Risk', 'Alpha', 'Trynor', 'Prob_Loss', 'Prob_Win', 'Odds_Winning',
                           'RSI Trend', 'MACD Trend', 'Trend'])

# Download NSE Index Data
data = pd.DataFrame()
data['^NSEI'] = yf.download('^NSEI', period="3y").Close
market_returns = np.log(data['^NSEI'] / data['^NSEI'].shift(1))
mr = market_returns.mean() * 252 * 100
market_var = market_returns.var() * 252

# Define RSI Calculation
def calculate_rsi(series, period=14):
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))

    # Classify RSI Trend
    if rsi.iloc[-1] > 60:
        return "Bullish"
    elif rsi.iloc[-1] < 40:
        return "Bearish"
    else:
        return "Neutral"

# Define MACD Calculation
def calculate_macd(series, short_window=12, long_window=26, signal_window=9):
    short_ema = series.ewm(span=short_window, adjust=False).mean()
    long_ema = series.ewm(span=long_window, adjust=False).mean()
    macd = short_ema - long_ema
    signal = macd.ewm(span=signal_window, adjust=False).mean()
    macd_hist = macd - signal

    # Classify MACD Trend
    if macd_hist.iloc[-1] > 0:
        return "Bullish"
    elif macd_hist.iloc[-1] < 0:
        return "Bearish"
    else:
        return "Neutral"

# Iterate over each ticker
for t in tickers:
    try:
        data[t] = yf.download(t, period="3y").Close
    except:
        continue

    sec_returns = np.log(data / data.shift(1))
    cov = sec_returns.cov() * 252
    cov_with_market = cov[t][0]

    dr = data[t].pct_change()
    dcum = (1 + dr).cumprod()
    n = len(data) / 252
    CAGR = ((dcum[-1]) ** (1 / n) - 1) * 100

    droll = dcum.cummax()
    ddw = droll - dcum
    ddmax = ddw / droll
    max_dd = (ddmax.max()) * 100

    returns = np.log(data[t] / data[t].shift(1))
    MSFT_beta = cov_with_market / market_var
    MSFT_er = returns.mean() * 252
    Sharp = (MSFT_er - 0.05) / (sec_returns[t].std() * 252 ** 0.5)

    vols = returns.std() * 252 ** 0.5 * 100
    annual_returns = returns.mean() * 252 * 100
    alpha = annual_returns - (7 + MSFT_beta * (mr - 7))
    CV = (vols / annual_returns) * 100

    # Monte Carlo Value at Risk (VaR)
    S = 100000  # Investment amount
    c = 0.95  # Confidence level
    n = 730  # 1 day
    iterations = 100000  # Number of paths
    mu = np.mean(returns)
    sigma = np.mean(returns.std())
    tr = (annual_returns - 7) / MSFT_beta
    model = ValueAtRiskMonteCarlo(S, mu, sigma, c, n, iterations)
    val = model.simulation()

    # Probability Calculations
    mean_return = annual_returns
    std_return = vols
    z_score = (0 - mean_return) / std_return
    probability_of_loss = stats.norm.cdf(z_score) * 100
    probability_of_winning = (1 - stats.norm.cdf(z_score)) * 100
    odds_of_winning = (probability_of_winning / probability_of_loss) if probability_of_loss > 0 else np.inf

    # Trend Analysis Indicators
    rsi_trend = calculate_rsi(data[t])
    macd_trend = calculate_macd(data[t])
    
    # Moving Average Crossover Trend
    dma_50 = data[t].rolling(window=50).mean().iloc[-1]
    dma_100 = data[t].rolling(window=100).mean().iloc[-1]

    if dma_50 > dma_100:
        trend = "Bullish"
    elif dma_50 < dma_100:
        trend = "Bearish"
    else:
        trend = "Neutral"

    # Store results in DataFrame
    df.loc[len(df.index)] = [t, cov_with_market, market_var, MSFT_beta, vols, annual_returns, CV, Sharp, CAGR, max_dd,
                             val, alpha, tr, probability_of_loss, probability_of_winning, odds_of_winning,
                             rsi_trend, macd_trend, trend]

# Display final DataFrame
```

```python
def t_stat_corr(df, x, y):
    """Returns the t-statistic for the correlation between 2 variables"""
    corr_xy = np.corrcoef(df[x], df[y])[0][1]
    n = len(df)
    
    t_stat = (corr_xy * np.sqrt(n - 2)) / np.sqrt(1 - corr_xy ** 2)
    
    return t_stat
```

```python
import numpy as np
import pandas as pd
import ta
import torch
import torch.nn as nn
import torch.optim as optim
import yfinance as yf

# Define the Actor network
class Actor(nn.Module):
    def __init__(self, state_size, action_size):
        super(Actor, self).__init__()
        self.fc1 = nn.Linear(state_size, 128)
        self.fc2 = nn.Linear(128, action_size)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.softmax(self.fc2(x), dim=-1)
        return x

# Define the Critic network
class Critic(nn.Module):
    def __init__(self, state_size):
        super(Critic, self).__init__()
        self.fc1 = nn.Linear(state_size, 128)
        self.fc2 = nn.Linear(128, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# Function to calculate metrics
def calculate_metrics(ticker_data, risk_free_rate):
    ticker_data['volatility'] = ticker_data['Close'].pct_change().rolling(window=252).std() * np.sqrt(252)
    ticker_data['returns'] = ticker_data['Close'].pct_change()
    ticker_data['cagr'] = (1 + ticker_data['returns'].mean()) ** 252 - 1
    ticker_data['sharpe_ratio'] = (ticker_data['returns'].mean() - risk_free_rate) / ticker_data['returns'].std() * np.sqrt(252)
    ticker_data['VaR'] = ticker_data['returns'].quantile(0.05)
    ticker_data['cv_ratio'] = ticker_data['returns'].std() / ticker_data['returns'].mean()

# Training the Actor-Critic agent
def actor_critic_train(data, alpha=0.01, gamma=0.99, episodes=1000, max_steps=50, risk_free_rate=0.01):
    state_size = 5
    action_size = len(data)
    
    actor = Actor(state_size, action_size)
    critic = Critic(state_size)
    
    actor_optimizer = optim.Adam(actor.parameters(), lr=alpha)
    critic_optimizer = optim.Adam(critic.parameters(), lr=alpha)
    
    tickers = list(data.keys())
    
    for episode in range(episodes):
        state = np.random.randint(0, state_size, size=(state_size,))
        for step in range(max_steps):
            state_tensor = torch.FloatTensor(state)
            action_probs = actor(state_tensor)
            action = np.random.choice(len(tickers), p=action_probs.detach().numpy())
            ticker = tickers[action]
            ticker_data = data[ticker]
            if ticker_data.empty or len(ticker_data) < 2:
                continue
            
            calculate_metrics(ticker_data, risk_free_rate)
            
            reward = 0
            if ticker_data['volatility'].iloc[-1] < 0.2:
                reward += 1
            if ticker_data['cagr'].iloc[-1] > 0.1:
                reward += 1
            if ticker_data['VaR'].iloc[-1] > -0.05:
                reward += 1
            if ticker_data['sharpe_ratio'].iloc[-1] > 1:
                reward += 1
            if ticker_data['cv_ratio'].iloc[-1] < 1:
                reward += 1
            
            new_state = np.random.randint(0, state_size, size=(state_size,))
            new_state_tensor = torch.FloatTensor(new_state)
            
            # Critic update
            value = critic(state_tensor)
            next_value = critic(new_state_tensor)
            target = reward + gamma * next_value
            critic_loss = (target - value).pow(2).mean()
            critic_optimizer.zero_grad()
            critic_loss.backward(retain_graph=True)
            critic_optimizer.step()
            
            # Actor update
            advantage = (target - value).detach()  # Detach advantage from the graph to avoid modifying it in place
            actor_loss = -torch.log(action_probs[action]) * advantage
            actor_optimizer.zero_grad()
            actor_loss.backward()
            actor_optimizer.step()
            
            state = new_state
    
    return actor, tickers

# Evaluate the trained Actor-Critic agent
def actor_critic_evaluate(actor, tickers, new_data, risk_free_rate=0.01):
    state_size = 5
    rewards = np.zeros(len(tickers))
    
    for action, ticker in enumerate(tickers):
        ticker_data = new_data[ticker]
        if ticker_data.empty or len(ticker_data) < 2:
            continue
        
        calculate_metrics(ticker_data, risk_free_rate)
        
        reward = 0
        if ticker_data['volatility'].iloc[-1] < 0.2:
            reward += 1
        if ticker_data['cagr'].iloc[-1] > 0.1:
            reward += 1
        if ticker_data['VaR'].iloc[-1] > -0.05:
            reward += 1
        if ticker_data['sharpe_ratio'].iloc[-1] > 1:
            reward += 1
        if ticker_data['cv_ratio'].iloc[-1] < 1:
            reward += 1
        
        rewards[action] = reward
    
    best_tickers = [tickers[i] for i in np.argsort(rewards)[-14:][::-1]]
    return best_tickers

# Example usage

# Training data


# Train the Actor-Critic agent
actor, tickers = actor_critic_train(data)



# Evaluate the trained Actor-Critic agent on new data
selected_stocks = actor_critic_evaluate(actor, tickers1, data1)
print(selected_stocks)
```

```python
import numpy as np
import pandas as pd
import yfinance as yf
from hmmlearn.hmm import GaussianHMM
from sklearn.preprocessing import StandardScaler
from ta.momentum import RSIIndicator
from ta.trend import MACD

# Load tickers from CSV
dfticker = pd.read_csv("D:\\stocks\\QGRMAY.csv", index_col=False)
tickers = dfticker['Ticker'].astype(str) + ".NS"

# Download stock data
data_raw = yf.download(tickers.tolist(), period="3y", group_by="ticker")

# Preprocess data
data = {}
for t in tickers:
    try:
        if 'Close' in data_raw[t] and not data_raw[t]['Close'].dropna().empty:
            df = data_raw[t][['Close']]
            df['Returns'] = df['Close'].pct_change()
            df['200DMA'] = df['Close'].rolling(window=200, min_periods=1).mean()
            df['RSI'] = RSIIndicator(df['Close']).rsi()
            df['MACD'] = MACD(df['Close']).macd()
            df.dropna(inplace=True)
            data[t] = df
    except KeyError:
        continue

tickers = list(data.keys())  # Update valid tickers

# Calculate financial metrics
def calculate_metrics(df, risk_free_rate=0.01):
    df = df.copy()
    df['Volatility'] = df['Returns'].std() * np.sqrt(252)
    df['CAGR'] = (1 + df['Returns'].mean()) ** 252 - 1
    df['Sharpe Ratio'] = (df['Returns'].mean() * 252 - risk_free_rate) / df['Returns'].std()
    df['Sortino Ratio'] = (df['Returns'].mean() * 252 - risk_free_rate) / df[df['Returns'] < 0]['Returns'].std()
    df['Max Drawdown'] = (df['Close'] / df['Close'].cummax() - 1).min()
    return df.dropna()

# Q-learning Training
def q_learning_train(data, alpha=0.1, gamma=0.9, epsilon=0.1, episodes=100, max_steps=50):
    tickers = list(data.keys())
    n_tickers = len(tickers)
    q_table = np.zeros((5, n_tickers))

    for episode in range(episodes):
        state = np.random.randint(0, 5)
        for step in range(max_steps):
            action = np.random.randint(0, n_tickers) if np.random.rand() < epsilon else np.argmax(q_table[state])
            ticker = tickers[action]
            df = data[ticker]
            
            if df.empty or len(df) < 2:
                continue

            df = calculate_metrics(df)

            reward = (
                (df['Volatility'].iloc[-1] < 0.2) +
                (df['CAGR'].iloc[-1] > 0.1) +
                (df['Sharpe Ratio'].iloc[-1] > 1) +
                (df['Sortino Ratio'].iloc[-1] > 1) +
                (df['Max Drawdown'].iloc[-1] > -0.2)
            )

            new_state = np.random.randint(0, 5)
            q_table[state, action] = q_table[state, action] + alpha * (reward + gamma * np.max(q_table[new_state]) - q_table[state, action])
            state = new_state

    return q_table, tickers

# Hidden Markov Model for Trend Prediction
def hmm_predict(df):
    df = df[['Returns', 'RSI', 'MACD']].dropna()
    scaler = StandardScaler()
    X = scaler.fit_transform(df)

    model = GaussianHMM(n_components=2, covariance_type="diag", n_iter=1000)
    model.fit(X)
    states = model.predict(X)

    trend = "Bullish" if np.mean(states) > 0.5 else "Bearish"
    return trend

# Q-learning Evaluation
def q_learning_evaluate(q_table, tickers, data):
    selected_stocks = []
    
    for i, ticker in enumerate(tickers):
        if ticker not in data:
            continue
        
        df = data[ticker]
        if df.empty or len(df) < 2:
            continue
        
        df = calculate_metrics(df)
        
        # Compute Q-value Score
        score = np.mean(q_table[:, i]) 
        
        if score > 0.5:
            trend = hmm_predict(df)  # Predict trend using HMM
            selected_stocks.append((ticker, score, trend))
    
    selected_stocks = sorted(selected_stocks, key=lambda x: x[1], reverse=True)
    
    return selected_stocks

# Load evaluation stock data
#dfticker1 = pd.read_csv("D:\\stocks\\HOLDINGSMARCH25.csv", index_col=False)
tickers1 = newtick
#tickers1 = (dfticker1['Instrument'].astype(str) + ".NS").tolist()
data_raw1 = yf.download(tickers1, period="3y", group_by="ticker")

data1 = {}
for t in tickers1:
    try:
        if 'Close' in data_raw1[t] and not data_raw1[t]['Close'].dropna().empty:
            df = data_raw1[t][['Close']]
            df['Returns'] = df['Close'].pct_change()
            df['200DMA'] = df['Close'].rolling(window=200, min_periods=1).mean()
            df['RSI'] = RSIIndicator(df['Close']).rsi()
            df['MACD'] = MACD(df['Close']).macd()
            df.dropna(inplace=True)
            data1[t] = df
    except KeyError:
        continue

tickers1 = list(data1.keys())

# Train & Evaluate
q_table, tickers = q_learning_train(data)
selected_stocks = q_learning_evaluate(q_table, newtick, data1)

# Display Selected Stocks
for stock, score, trend in selected_stocks:
    print(f"Stock: {stock}, Score: {score:.2f}, Trend: {trend}")
```

```python
import numpy as np
import pandas as pd
import yfinance as yf
from statsmodels.tsa.stattools import coint

# Download historical data for multiple stocks
tickers = newtick
data = yf.download(tickers, start='2021-01-01')['Close']
data = data.dropna()
# Calculate correlation matrix
correlation_matrix = data.corr()

# Identify pairs with high correlation
threshold = 0.8
correlated_pairs = [(col1, col2) for col1 in correlation_matrix.columns for col2 in correlation_matrix.columns 
                    if col1 != col2 and correlation_matrix[col1][col2] > threshold]

# Check for cointegration
for pair in correlated_pairs:
    score, p_value, _ = coint(data[pair[0]], data[pair[1]])
    if p_value < 0.05:  # Significance level
        print(f"Cointegrated pair: {pair}, p-value: {p_value:.4f}")
   
```
